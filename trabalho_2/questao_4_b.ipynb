```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the function
def f(x):
  return (1-x**2)**(3/2)

N=100000
x=np.random.uniform(0,1,N)
```

```python
##### Monte Carlo ########
integral=sum(f(x))/N

# Create the x-axis values
xplot = np.linspace(0, 1, 100)


# Plot the function
plt.plot(xplot, f(xplot))

# Fill the area under the curve
plt.fill_between(xplot, f(xplot), color='lightblue')

# Add the label with the area value
plt.text(0.5, 0.3, f'Area = {integral:.3f}', fontsize=12)

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')

# Show the plot
plt.show()
```

```python
##### Amostragem por import√¢ncia ######

def G(x):
  return -np.log(1 - x * (1 - np.exp(-1)))

def g(x):
  return np.exp(-x)/(1-np.exp(-1))

#Gerar as amostras de g(x)
Y = G(x)


#calcular a integral
integral = np.sum( f(Y)/g(Y) ) / N

#plotar o grafico
xplot = np.linspace(0, 1, N)
curve_y = f(xplot)
curve_g = g(xplot)
plt.plot(xplot, curve_y, color='black')
plt.hist(Y, color='red', bins=100, density=True)
plt.fill_between(xplot, curve_y, color='blue')
plt.text(0.5, 1.2, f'Integral = {integral:.4f}', fontsize=12, color='red')
# plt.xlim(0, 1)
# plt.ylim(0, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
