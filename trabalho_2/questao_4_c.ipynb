```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the function
def f(x):
  return x*(1+x**2)**(-2)

a=0
b=8

N=100000
x=np.random.uniform(a,b,N)
```

```python
##### Monte Carlo ########
integral=sum(f(x))/N * (b-a)

# Create the x-axis values
xplot = np.linspace(a, b, 100)


# Plot the function
plt.plot(xplot, f(xplot))

# Fill the area under the curve
plt.fill_between(xplot, f(xplot), color='lightblue')

# Add the label with the area value
plt.text(4, 0.1, f'Area = {integral:.3f}', fontsize=12)

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')

# Show the plot
plt.show()
```

```python
##### Amostragem por importância ######
x=np.random.uniform(0,1,N)

def G(x):
  return -2.5 * np.log(1-x)

def g(x):
  return 0.4 * np.exp(-x * 0.4)

#Gerar as amostras de g(x)
Y = G(x)

#calcular a integral
integral = np.sum( f(Y)/g(Y) ) / N

#plotar o grafico
xplot = np.linspace(a, b, N)
curve_y = f(xplot)
curve_g = g(xplot)
plt.hist(Y, color='red', bins=100, density=True)
plt.plot(xplot, curve_y, color='black')
plt.fill_between(xplot, curve_y, color='blue')
plt.text(4, 0.2, f'Integral = {integral:.4f}', fontsize=12, color='red')
# plt.xlim(0, 1)
# plt.ylim(0, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
