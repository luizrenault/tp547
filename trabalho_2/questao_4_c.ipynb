```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the function
def f(x):
  return np.exp(x + x**2)

a=-2
b=2

N=100000
x=np.random.uniform(a,b,N)
```

```python
##### Monte Carlo ########
integral=sum(f(x))/N * (b-a)

# Create the x-axis values
xplot = np.linspace(a, b, 100)


# Plot the function
plt.plot(xplot, f(xplot))

# Fill the area under the curve
plt.fill_between(xplot, f(xplot), color='lightblue')

# Add the label with the area value
plt.text(0.5, 200, f'Area = {integral:.3f}', fontsize=12)

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')

# Show the plot
plt.show()
```
