```python
import numpy as np
import matplotlib.pyplot as plt

n=100000
X=np.array([])

for i in range(n):
    x1=np.random.uniform(-1,1)
    x2=np.random.uniform(0,1.5)

    while (x2 >= 1.5*x1**2):
        x1=np.random.uniform(-1,1)
        x2=np.random.uniform(0,1.5)
    X = np.append(X, x1)

x=np.arange(-1,1,0.01)
fx=1.5*x**2

print(X)
plt.plot(x,fx)
plt.hist(X, bins=100, density=True)
plt.show()
```
